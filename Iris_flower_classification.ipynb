```python
#importing KNeighborsClassifier for performing knn,from sklearn library 
from sklearn.neighbors import KNeighborsClassifier
#importing iris dataset from sklearn library
from sklearn.datasets import load_iris
#importing metrics from sklearn module for accuracy calculation
from sklearn import metrics
#loading the dataset
iris=load_iris()
#all about dataset
print('The names of the dataset columns:\n',iris['feature_names'])
print('The names of target classes:\n',iris['target_names'])
print('The description of the dataset:\n',iris['DESCR'])
```

```python
X = iris.data#features
y= iris.target #target variables
from matplotlib import pyplot as plt
# The indices of the features that we are plotting
x_index = 0
y_index = 1
#visualizing the data 
# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])
plt.figure(figsize=(4,4))
plt.scatter(iris.data[:, x_index], iris.data[:, y_index], c=iris.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.tight_layout()
plt.show()
```

```python
#splitting the dataset into trainig and testing sets
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(X,y, test_size= 0.25, random_state=0)
#performing knn classification
knn = KNeighborsClassifier(n_neighbors=3)
#fitting the data
knn.fit(X,y)
#predicted the data
y_pred= knn.predict(x_test)
```

```python
#printing consfusion matrix
from sklearn.metrics import confusion_matrix
print('Confusion matrix\n',confusion_matrix(y_test, y_pred))

#printing the classification report
from sklearn.metrics import classification_report
print('Classification report\n:',classification_report(y_test, y_pred))

#printing the predicted values and accuracy
print('predicted values:\n',y_pred)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
